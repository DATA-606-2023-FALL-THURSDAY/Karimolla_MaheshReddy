```python
Data = pd.read_csv('data.csv')
```

```python
Data.drop(['Unnamed: 27','TAXI_OUT',
       'WHEELS_OFF',
       'WHEELS_ON', 'TAXI_IN','CANCELLATION_CODE'],inplace =True, axis =1)
```

```python
Data['FL_DATE'] = pd.to_datetime(Data['FL_DATE'])

# extract year, month, and day from 'FL_DATE' column
Data['year'] = Data['FL_DATE'].dt.year
Data['month'] = Data['FL_DATE'].dt.month
Data['day'] = Data['FL_DATE'].dt.day

# print the updated dataframe
Data.head(5)
```

```python
data = Data [['year', 'month', 'day','OP_CARRIER', 'OP_CARRIER_FL_NUM', 'ORIGIN', 'DEST',
       'CRS_DEP_TIME', 'DEP_TIME', 'DEP_DELAY', 'CRS_ARR_TIME', 'ARR_TIME',
       'ARR_DELAY', 'CANCELLED', 'DIVERTED', 'CRS_ELAPSED_TIME',
       'ACTUAL_ELAPSED_TIME', 'AIR_TIME', 'DISTANCE', 'CARRIER_DELAY',
       'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]
```

```python
data = data.astype({"year": 'int16', "month": 'int16',
                    'day':'int16','OP_CARRIER': 'category',
                    'OP_CARRIER_FL_NUM': 'int16', 'ORIGIN':'category',
                    'DEST':'category','CRS_DEP_TIME': 'float16',
                    'DEP_TIME': 'float16', 'DEP_DELAY':'float16',
                    'CRS_ARR_TIME':'float16', 'ARR_TIME':'float16',
       'ARR_DELAY': 'float16', 'CANCELLED':'float16',
                    'DIVERTED':'float16', 'CRS_ELAPSED_TIME' :'float16',
       'ACTUAL_ELAPSED_TIME': 'float16', 'AIR_TIME':'float16',
                    'DISTANCE': 'float16', 'CARRIER_DELAY':'float16',
       'WEATHER_DELAY':'float16', 'NAS_DELAY':'float16',
                    'SECURITY_DELAY':'float16', 'LATE_AIRCRAFT_DELAY':'float16'})
```

```python
data.info()
```

```python
Final_data = data.copy()
```

```python
Final_data
```

```python
Final_data.isnull().sum()
```

```python
year = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
CARRIER_DELAY = (5.28/100)
WEATHER_DELAY = (0.57/100)
NAS_DELAY = (5.88/100)
SECURITY_DELAY = (0.04/100)
LATE_AIRCRAFT_DELAY = (0.24/100)
for i in range(len(year)):
    Final_data[Final_data['year'] == year[i]]
    Final_data['CARRIER_DELAY'] = Final_data['DEP_DELAY'] * CARRIER_DELAY
    Final_data['WEATHER_DELAY'] = Final_data['DEP_DELAY'] * WEATHER_DELAY
    Final_data['NAS_DELAY'] = Final_data['DEP_DELAY'] * NAS_DELAY
    Final_data['SECURITY_DELAY'] = Final_data['DEP_DELAY'] * SECURITY_DELAY
    Final_data['LATE_AIRCRAFT_DELAY'] = Final_data['DEP_DELAY'] * LATE_AIRCRAFT_DELAY
```

```python
Final_data = Final_data.dropna()
```
